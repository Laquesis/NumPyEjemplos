EJEMPLO DIAGRAMA DE DISPERSIÓN

Este código genera un conjunto de datos (x y z), organiza los datos en un DataFrame y luego los grafica en un diagrama de dispersión (scatter plot). La visualización resultante muestra los puntos (x, z) en rojo, con etiquetas en los ejes y una leyenda que los identifica como "Medido".

```python
# Importamos las bibliotecas necesarias
import numpy as np                # Biblioteca para manejo de arreglos y funciones matemáticas
import pandas as pd               # Biblioteca para manejo y análisis de datos en forma de tablas
import matplotlib.pyplot as plt   # Biblioteca para visualización de datos

# Definimos la cantidad de puntos de datos
n = 31

# Generamos una serie de valores equiespaciados entre 0 y 3 (31 puntos en total)
x = np.linspace(0, 3, n)

# Creamos un arreglo con datos específicos (mediciones o valores a graficar)
z = np.array([1.89, -0.12, -0.32, 2.11, -0.25, 1.01, 0.17, 2.75, 2.01, 5.5,
              0.87, 3.31, 2.29, 2.73, 3.67, 3.92, 4.29, 5.27, 3.85, 4.26,
              5.75, 5.82, 6.36, 9.13, 7.61, 9.52, 9.53, 12.49, 12.29, 13.7, 14.12])

# Creamos un DataFrame con las columnas 'x' y 'z' que contienen los datos de x y z respectivamente
data = pd.DataFrame(np.vstack((x, z)).T, columns=['x', 'z'])

# Graficamos los puntos de datos (x, z) como puntos rojos
plt.plot(x, z, 'ro', label='Medido')    # 'ro' indica puntos rojos, y 'label' define la leyenda

# Añadimos etiquetas a los ejes y mostramos la leyenda
plt.xlabel('x')                         # Etiqueta del eje x
plt.ylabel('z')                         # Etiqueta del eje z
plt.legend()                            # Muestra la leyenda en el gráfico

# Mostramos el gráfico
plt.show()
```

El aprendizaje automático corresponde a algoritmos de computador y modelos estadísticos que se basan en patrones e inferencias. Estos algoritmos realizan una tarea específica sin instrucciones explícitas. Los modelos de regresión con aprendizaje automático pueden ser tan simples como una regresión lineal o tan complejos como el aprendizaje profundo (deep learning). Este tutorial indica varios métodos de regresión con scikit-learn.



```python
# %matplotlib inline permite mostrar gráficos en línea en entornos como Jupyter Notebook.
%matplotlib inline

# Importamos Axes3D para crear gráficos en 3D
from mpl_toolkits.mplot3d import Axes3D

# Importamos la biblioteca math para funciones matemáticas avanzadas
import math

# Definimos la función f(x, y), que calcula un valor z en función de x e y
def f(x, y):
    return 2 * math.cos(x) * y + x * math.cos(y) - 3 * x * y

# Definimos la cantidad de puntos que generaremos
n = 500

# Generamos valores aleatorios para x e y en el rango de [-1, 1)
x = (np.random.rand(n) - 0.5) * 2.0
y = (np.random.rand(n) - 0.5) * 2.0

# Creamos un arreglo vacío para almacenar los valores de z calculados
z = np.empty_like(x)

# Calculamos el valor de z para cada par de valores (x, y) usando la función f(x, y)
for i in range(n):
    z[i] = f(x[i], y[i])

# Creamos un DataFrame con las columnas 'x', 'y' y 'z' que contienen los datos generados
data = pd.DataFrame(np.vstack((x, y, z)).T, columns=['x', 'y', 'z'])

# Configuramos la figura para el gráfico 3D
fig = plt.figure()

# Añadimos un subplot 3D para graficar los datos
ax = fig.add_subplot(111, projection='3d')

# Generamos un gráfico de dispersión en 3D con los valores de (x, y, z)
# Usamos un mapa de colores (cmap='plasma') para colorear los puntos en función de los valores de z
ax.scatter(x, y, z, c=z, cmap='plasma')

# Mostramos el gráfico en pantalla
plt.show()
```

Explicación:
Creación de los Datos Originales: Se define un conjunto de puntos de datos (x, y).
Organización en un DataFrame: Los datos se almacenan en un DataFrame para su análisis o visualización si fuera necesario.
Interpolación con scipy: Creamos una función de interpolación f_interpolacion que usa la opción cúbica (kind='cubic'), que crea una curva suave.
Interpolación en Nuevos Puntos: Definimos un rango de puntos x_nuevos para obtener valores de y en posiciones intermedias mediante la función de interpolación.
Visualización: Se grafican los puntos originales junto con la curva interpolada.

EJEMPLO INTERPOLACION CON UNA RED NEURONAL

```python
# Importamos las librerías necesarias
from gekko import brain
import numpy as np
import matplotlib.pyplot as plt

# Generamos valores para x entre 0 y 2π y calculamos y como el seno de x
x = np.linspace(0.0, 2 * np.pi)
y = np.sin(x)

# Creamos una instancia de Brain, la clase de redes neuronales de Gekko
# El parámetro `remote=False` indica que el cálculo se hará de forma local
b = brain.Brain(remote=False)

# Definimos la estructura de la red neuronal:
b.input_layer(1)         # Capa de entrada con 1 neurona (para entrada de 1D)
b.layer(linear=2)        # Primera capa oculta, lineal, con 2 neuronas
b.layer(tanh=2)          # Segunda capa oculta, función de activación tangente hiperbólica (tanh) con 2 neuronas
b.layer(linear=2)        # Tercera capa oculta, lineal, con 2 neuronas
b.output_layer(1)        # Capa de salida con 1 neurona (para salida de 1D)

# Entrenamos la red neuronal con los datos de entrada (x) y salida (y)
# `disp=False` evita mostrar mensajes de entrenamiento en pantalla
b.learn(x, y, disp=False)

# Generamos nuevos puntos `xp` para evaluar el modelo en un rango mayor (-2π a 4π)
xp = np.linspace(-2 * np.pi, 4 * np.pi, 100)

# Utilizamos la red neuronal entrenada para predecir los valores `yp` en los nuevos puntos `xp`
yp = b.think(xp)

# Graficamos los resultados:
plt.figure()
plt.plot(x, y, 'bo', label='medida (etiqueta)')   # Puntos originales en azul (sin(x))
plt.plot(xp, yp[0], 'r-', label='modelo')         # Predicción de la red en rojo
plt.legend()                                      # Mostramos la leyenda
plt.show()    
```

EJEMPLO INTERPOLACIÓN

```python
# Importamos las bibliotecas necesarias
#numpy
import numpy as np
#pandas
import pandas as pd
#Biblioteca graficas
import matplotlib.pyplot as plt
#SCIPY
from scipy.interpolate import interp1d

# Creamos datos iniciales (x e y) para interpolar
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 0.8, 0.9, 0.1, -0.8, -1.0])

# Organizamos los datos en un DataFrame para facilitar su manejo
data = pd.DataFrame({'x': x, 'y': y})

# Creamos una función de interpolación lineal usando scipy.interpolate.interp1d
# El parámetro 'kind' define el tipo de interpolación (puede ser 'linear', 'quadratic', 'cubic', etc.)
f_interpolacion = interp1d(x, y, kind='cubic')

# Definimos nuevos puntos en el eje x para interpolar entre los datos originales
x_nuevos = np.linspace(0, 5, 50)  # 50 puntos entre 0 y 5
y_nuevos = f_interpolacion(x_nuevos)  # Interpolamos en estos nuevos puntos

# Graficamos los puntos originales y los puntos interpolados
plt.plot(x, y, 'o', label='Puntos originales')        # Datos originales
plt.plot(x_nuevos, y_nuevos, '-', label='Interpolación (cúbica)')  # Interpolación
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolación usando scipy, numpy y pandas')
plt.show()
```
